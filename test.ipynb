简易文件批量处理

```python
import os
import shutil
dest_dir_path="dir_test"
walk_dir_path="walk_test"
loc_path=os.getcwd()
filepath=os.path.join(loc_path,dest_dir_path)
filepath
```

```python
def RemoveDir(filepath):
    '''
    如果文件夹不存在就创建，如果文件夹存在就清空！
    '''
    
    if not os.path.exists(filepath):
        os.mkdir(filepath)
    else:
        shutil.rmtree(filepath)
        os.mkdir(filepath)
```

```python
#生成测试数据
for i in range(1000):
    file=open(os.path.join(filepath,str(i)+".txt"),"w")
    file.close()
    
```

```python
#清空测试数据
RemoveDir(filepath)
```
